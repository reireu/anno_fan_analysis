```python
import matplotlib.pyplot as plt
import pandas as pd

# Twitterのデータを読み込む
twitter_data = pd.read_csv('followers_data.csv', parse_dates=['DATE'])

# YouTubeのデータを読み込む
youtube_data = pd.read_csv('subscribers_data.csv', parse_dates=['DATE'])

# グラフの設定
plt.figure(figsize=(12, 6))

# 左側のプロット（Twitterフォロワー数）
plt.subplot(1, 2, 1)
plt.plot(twitter_data['DATE'], twitter_data['FOLLOWERS'], marker='o', linestyle='-', color='b', label='Twitter フォロワー')
plt.title('Twitterフォロワー数の推移')
plt.xlabel('日付')
plt.ylabel('フォロワー数')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

# 右側のプロット（YouTubeサブスクライバー数）
plt.subplot(1, 2, 2)
plt.plot(youtube_data['DATE'], youtube_data['SUBSCRIBERS'], marker='o', linestyle='-', color='g', label='YouTube サブスクライバー')
plt.title('YouTubeサブスクライバー数の推移')
plt.xlabel('日付')
plt.ylabel('サブスクライバー数')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

# レイアウト調整と表示
plt.tight_layout()
plt.show()
```

```python
# グラフの設定
fig, ax1 = plt.subplots(figsize=(20, 12))

# Twitterのフォロワー数の折れ線グラフ
ax1.plot(twitter_data['DATE'], twitter_data['FOLLOWERS'], marker='o', linestyle='-', color='#7fb3d5', label='Twitter フォロワー')
ax1.set_xlabel('日付')
ax1.set_ylabel('Twitterフォロワー数', color='black')
ax1.tick_params(axis='y', labelcolor='black')
ax1.grid(True)
ax1.legend(loc='upper left')

# YouTubeの登録者数の折れ線グラフ
ax2 = ax1.twinx()
ax2.plot(youtube_data['DATE'], youtube_data['SUBSCRIBERS'], marker='o', linestyle='-', color='#b2df8a', label='YouTube登録者数')
ax2.set_ylabel('YouTube登録者数', color='black')
ax2.tick_params(axis='y', labelcolor='black')
ax2.legend(loc='upper right')

# グラフのタイトル
plt.title('Twitterフォロワー数とYouTube登録者数の推移')

# レイアウト調整と表示
plt.tight_layout()
plt.show()
```
