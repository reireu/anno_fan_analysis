```python
import matplotlib.pyplot as plt
import pandas as pd

# Twitter
twitter_data = pd.read_csv('followers_data.csv', parse_dates=['DATE'])

# YouTube
youtube_data = pd.read_csv('subscribers_data.csv', parse_dates=['DATE'])

# 
plt.figure(figsize=(12, 6))

# Twitter
plt.subplot(1, 2, 1)
plt.plot(twitter_data['DATE'], twitter_data['FOLLOWERS'], marker='o', linestyle='-', color='b', label='Twitter ')
plt.title('Twitter')
plt.xlabel('')
plt.ylabel('')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

# YouTube
plt.subplot(1, 2, 2)
plt.plot(youtube_data['DATE'], youtube_data['SUBSCRIBERS'], marker='o', linestyle='-', color='g', label='YouTube ')
plt.title('YouTube')
plt.xlabel('')
plt.ylabel('')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

# 
plt.tight_layout()
plt.show()
```

```python
# 
fig, ax1 = plt.subplots(figsize=(20, 12))

# Twitter
ax1.plot(twitter_data['DATE'], twitter_data['FOLLOWERS'], marker='o', linestyle='-', color='#7fb3d5', label='Twitter ')
ax1.set_xlabel('')
ax1.set_ylabel('Twitter', color='black')
ax1.tick_params(axis='y', labelcolor='black')
ax1.grid(True)
ax1.legend(loc='upper left')

# YouTube
ax2 = ax1.twinx()
ax2.plot(youtube_data['DATE'], youtube_data['SUBSCRIBERS'], marker='o', linestyle='-', color='#b2df8a', label='YouTube')
ax2.set_ylabel('YouTube', color='black')
ax2.tick_params(axis='y', labelcolor='black')
ax2.legend(loc='upper right')

# 
plt.title('TwitterYouTube')

# 
plt.tight_layout()
plt.show()
```
